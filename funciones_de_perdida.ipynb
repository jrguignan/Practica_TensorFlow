```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

# Tipos de Función de Pérdida - Regresión

```python
#valores de prueba
X = tf.linspace(-1., 1., 100)

target = tf.constant(0.)
```

Norma en L2

$$L2(y_r , y_p) = (y_r - y_p)^2 $$

```python
L2 = tf.math.square(target - X)

print(L2)
```

Norma L1 - Función de pérdidas absolutas

$$L1(y_r , y_p) = |y_r - y_p|  ^2 $$


Es mejor para los outliers, pero no es derivable

```python
L1 = tf.math.abs(target-X) 

print(L1)
```

Psudo-Haber

$$ PH(y_r , y_p) = \delta^2 \cdot \sqrt{1+\left(\frac{y_r - y_p}{\delta}\right)^2}-1 , \hspace{0.5cm} \delta > 0 $$

```python
delta = tf.constant(0.25, dtype='float32')

PHy = tf.multiply(tf.math.square(delta),tf.math.sqrt(tf.add(1.,tf.math.square((target-X)/delta)))-1. )

print(PHy)
```

```python
#Varilando el delta
delta = tf.constant(5.0, dtype='float32')

PHy_1 = tf.multiply(tf.math.square(delta),tf.math.sqrt(tf.add(1.,tf.math.square((target-X)/delta)))-1. )

print(PHy_1)
```

# Tipos de Función de Pérdida - Clasificación

```python
#Valores de prueba
X = tf.linspace(-3., 5., 100 )

target = tf.constant(1.0)
targets = tf.fill([100,], 1.)
```

Hinge - Función Bisagra
$$ HG(y_r, y_p) = max(0,1-y_r\cdot y_r) $$


```python
HG = tf.maximum(0., 1.0 - tf.multiply(target,X))
print(HG)
```

Entropía Cruzada - Fución Logística

$$ H(y_r, y_p) = -y_r\cdot log(y_p) - (1-y_r) \cdot log(1-y_p) $$

```python
X_entropy = -tf.multiply(target,tf.math.log(X))-tf.multiply(1.-target,tf.math.log(1.-X))

print(X_entropy)

#el logaritmo de un numero negativo no existe, los datos deben estar acotados
#por eso los nan
```

Entropía Cruzada - Sigmoide

$$ H(y_r, y_p) = -y_r\cdot log\left(\frac{1}{1+e^{-y_p}}\right) - (1-y_r) \cdot log\left(1-\frac{1}{1+e^{-y_p}}\right) $$

```python
x_sig = tf.nn.sigmoid_cross_entropy_with_logits(logits=X, labels=targets)

print(x_sig)
```

```python
#sigmoide ponderada

ponderacion = tf.constant(0.5)

x_sig_pond = tf.nn.weighted_cross_entropy_with_logits(logits=X, labels=targets, pos_weight=ponderacion)

print(x_sig_pond)
```

Softmax Cross Entropy

$$SM(z_k) = \frac{e^{z_k}}{\sum_{i=1}^n e^{z_k}} $$

```python
unscale_logi = tf.constant([[1., -3., 10. ]])
target_dist = tf.constant([0.1, 0.02, 0.88])

soft = tf.nn.softmax_cross_entropy_with_logits(logits=unscale_logi, labels=target_dist)

print(soft)
```

Sparse Softmax Cross Entropy

```python
unscale_logi = tf.constant([[1., -3., 10. ]])
#Se utiliza la posicion de mayor probabilidad
sparse_target_dist = tf.constant([2])

soft_cross = tf.nn.sparse_softmax_cross_entropy_with_logits(logits=unscale_logi, labels=sparse_target_dist)

print(soft_cross)
```
